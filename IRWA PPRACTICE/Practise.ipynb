```python
# Find and print duplicates directly from movies_df
duplicates = movies_df[movies_df.duplicated(subset='title', keep=False)]
print("Duplicates before removal:")
print(duplicates)

# Remove duplicates based on the 'title' column
movies_df = movies_df.drop_duplicates(subset='title', keep='first')

# Reset index if desired
movies_df.reset_index(drop=True, inplace=True)

# Display the DataFrame with duplicates removed
print("\nDataFrame after removing duplicates:")
print(movies_df)
```

```python
# Convert 'genres' and 'keywords' from JSON-like strings to lists of genre/keyword names.
def parse_json_col(json_col):
    if pd.isna(json_col):
        return []
    return [item['name'] for item in ast.literal_eval(json_col)]

movies_df['genres'] = movies_df['genres'].apply(parse_json_col)
movies_df['keywords'] = movies_df['keywords'].apply(parse_json_col)
movies_df['cast'] = movies_df['cast'].apply(parse_json_col)

```

```python
# Tokenize and vectorize the 'overview' column using TF-IDF.
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(movies_df['overview'])

# Store the TF-IDF sparse matrix directly in the DataFrame
#movies_df['overview_tokevec'] = [tfidf_matrix[i, :].toarray().flatten() for i in range(tfidf_matrix.shape[0])]

# Print the DataFrame
print(movies_df[['overview']])
```

```python
# Normalize 'popularity', 'vote_average', 'vote_count'
scaler = MinMaxScaler()

scaled_columns = ['popularity', 'vote_average', 'vote_count']
movies_df[scaled_columns] = scaler.fit_transform(movies_df[scaled_columns])
```

```python
# Convert 'release_date' to datetime and extract the year.
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')
movies_df['release_year'] = movies_df['release_date'].dt.year
```

```python
# Function to check for symbols (excluding common punctuation marks)
def check_for_symbols(column):
    pattern = re.compile(r'[^\w\s.,;:!?\'"\-()]')  
    return movies_df[column].apply(lambda x: bool(pattern.search(str(x))))

# Calculate the total number of titles with symbols
title_symbols_count = check_for_symbols('title').sum()
print(f"Total titles with symbols before dropping: {title_symbols_count}")

```

```python
# Drop rows that contain symbols in the title column
movies_df = movies_df[~check_for_symbols('title')]

# Final check to print the total titles with symbols after dropping
final_title_symbols_count = check_for_symbols('title').sum()
print(f"Total titles with symbols after dropping: {final_title_symbols_count}")
```

```python
# Function to remove spaces in both lists and strings
def remove_space(word):
    if isinstance(word, list):  # If it's a list, join elements into a string
        # Remove spaces within list elements, then join the list into a single string
        return ', '.join([item.replace(" ", "") for item in word])
    elif isinstance(word, str):  # If it's a string, directly remove spaces
        return word.replace(" ", "")
    else:
        return word  # In case it's neither, just return it as it is

# Apply the function to the relevant columns (excluding 'overview')
movies_df['cast'] = movies_df['cast'].apply(remove_space)
movies_df['genres'] = movies_df['genres'].apply(remove_space)
movies_df['keywords'] = movies_df['keywords'].apply(remove_space)

# Ensure no missing values (NaN) before concatenating
movies_df['genres'] = movies_df['genres'].fillna('')
movies_df['keywords'] = movies_df['keywords'].fillna('')
movies_df['cast'] = movies_df['cast'].fillna('')

# Create the 'tags' column by concatenating the cleaned 'genres', 'keywords', and 'cast' columns
movies_df['tags'] = movies_df['genres'] + ' ' + movies_df['keywords'] + ' ' + movies_df['cast']

# Display the resulting DataFrame with 'title' and 'tags' columns
print(movies_df[['title', 'tags']].head())
```

Collaborative filtering

```python
# Step 1: Load the ratings data
ratings = pd.read_csv('ratings_new.csv')


ratings = ratings[['userId','movie_id','rating']]
ratings.head()
```

```python
# Step 3: Merge the two datasets on movie_id
merged_df = pd.merge(ratings, movies_df, on='movie_id')

# Step 4: Check the merged dataset
print(merged_df.head())

# Save the merged dataset
merged_df.to_csv('new_merged_movies.csv', index=False)
```

```python
# Ensure movie_id is an integer
merged_df['movie_id'] = merged_df['movie_id'].astype(int)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Step 1: Pivot the data to get a user-movie matrix
user_movie_matrix = merged_df.pivot_table(index='userId', columns='movie_id', values='rating')

# Step 2: Normalize by subtracting the mean rating for each user
user_movie_matrix_mean = user_movie_matrix.subtract(user_movie_matrix.mean(axis=1), axis=0)

# Step 3: Compute cosine similarity between users
user_similarity = cosine_similarity(user_movie_matrix_mean.fillna(0))

# Convert the similarity matrix into a DataFrame
user_similarity_df = pd.DataFrame(user_similarity, index=user_movie_matrix.index, columns=user_movie_matrix.index)

def recommend_movies_collaborative(user_id, num_recommendations=5):
    # Get the similarity scores for the target user
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)

    # Movies the user has already watched
    user_watched_movies = user_movie_matrix.loc[user_id].dropna().index.tolist()

    # Debug: Print watched movies
    print(f"User {user_id} has watched movies: {user_watched_movies}")

    # Initialize an empty Series to store movie recommendations
    recommended_movies = pd.Series(dtype=float)

    # Loop through similar users to gather movies they rated highly
    for similar_user_id in similar_users.index[1:]:
        similar_user_ratings = user_movie_matrix.loc[similar_user_id].dropna()

        # Debug: Print similar user ratings
        print(f"Similar User {similar_user_id} ratings: {similar_user_ratings}")

        # Filter out movies the target user has already watched
        new_movies = similar_user_ratings[~similar_user_ratings.index.isin(user_watched_movies)]
        
        # Debug: Print new movies found
        print(f"New movies from User {similar_user_id} for User {user_id}: {new_movies.index.tolist()}")

        # Add these movies to the recommendation list
        recommended_movies = pd.concat([recommended_movies, new_movies])

        # Stop if we've gathered enough recommendations
        if len(recommended_movies) >= num_recommendations:
            break
    
    # Get the top N recommended movie IDs
    if not recommended_movies.empty:
        recommended_movie_ids = recommended_movies.sort_values(ascending=False).head(num_recommendations).index.tolist()
        
        # Step 5: Map the movie IDs to their titles
        recommended_movie_titles = merged_df.loc[merged_df['movie_id'].isin(recommended_movie_ids), 'title'].unique()
        return recommended_movie_titles, recommended_movie_ids  # Return both titles and IDs
    else:
        return [], []  # Return empty lists if no recommendations were found

# Example: Prompt user for ID and recommend 5 movies
user_id_input = int(input("Enter user ID: "))
recommended_titles, recommended_ids = recommend_movies_collaborative(user_id=user_id_input, num_recommendations=5)

print("Recommended movie titles for User", user_id_input, ":")
for title in recommended_titles:
    print(title)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

def evaluate_recommendations(movie, relevant_movies):
    # Get recommended movie titles
    recommended_movies = recommend(movie)
    
    # Check if the recommended_movies is None or empty
    if recommended_movies is None or len(recommended_movies) == 0:
        print("No recommendations available for the movie:", movie)
        return

    print("Recommended Movies:", recommended_movies)  # Debug: Show recommended movies

    # Create a set of all unique movie titles from both relevant and recommended lists
    all_movies = set(recommended_movies).union(set(relevant_movies))

    # Generate binary lists based on the presence in recommended and relevant movies
    y_true = [1 if movie in relevant_movies else 0 for movie in all_movies]
    y_pred = [1 if movie in recommended_movies else 0 for movie in all_movies]

    # Debugging prints
    print("Relevant Movies:", relevant_movies)  # Debug: Show relevant movies
    print("All Movies:", all_movies)  # Debug: Show all unique movies being compared
    print("Binary True Labels:", y_true)  # Debug: Show binary true labels
    print("Binary Predicted Labels:", y_pred)  # Debug: Show binary predicted labels

    # Handle case where no recommendations or relevant movies exist
    if len(recommended_movies) == 0 or len(relevant_movies) == 0:
        print("No recommendations or relevant movies found.")
        return

    # Calculate Precision, Recall, and F1-Score
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1-Score: {f1:.2f}")

# Example usage
relevant_movies = ['The Dark Knight', 'Batman Forever', 'Batman', 'Batman Begins']  # Replace with actual relevant movies
evaluate_recommendations('The Dark Knight Rises', relevant_movies)
```
